```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
plt.rcParams['savefig.dpi'] = 10
```

```python
df = pd.read_csv('Data/control_varying_delta_Pmax_0.1', delim_whitespace = True, header = None)
df.head(2)
```

```python
delta = df.loc[:,0]
controls = df.loc[:, 1:8]
controls_abs = controls.abs()
controls_sum = controls_abs.sum(axis=1)
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
#ax.plot(delta, controls, markevery=10000)
#ax.plot(delta, controls[0], label = "1")
ax.plot(delta, controls[1], label = "2", color='blue')
ax.plot(delta, controls[2], label = "3", color='green')
ax.plot(delta, controls[3], label = "4", color='orange')
ax.plot(delta, controls[4], label = "5", color='black')
ax.plot(delta, controls[5], label = "6", color='red')
ax.plot(delta, controls[6], label = "7", color='fuchsia')
ax.plot(delta, controls[7], label = "8", color='orange')
plt.legend(loc='upper right', prop={'size': 15}, framealpha=0.6)
ax.set_xlabel('$\delta$', fontsize=20, rotation=0, size = 26)
ax.set_ylabel('$P_{max} tanh( \delta \cdot \Theta_{i} )$', fontsize=18, rotation=90, size = 26)
plt.grid()
#plt.savefig('control_varying_delta_Pmax_0.1.png', bbox_inches='tight', dpi = 80)
plt.savefig('control_varying_delta_Pmax_01.png', bbox_inches='tight', dpi = 40)
plt.show()
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.plot(delta, controls_sum)
ax.set_xlabel('$\delta$', fontsize=20, rotation=0, size = 26)
ax.set_ylabel('$\sum_i | P_{max} tanh( \delta \cdot \Theta_{i} )| $', fontsize=18, rotation=90, size = 26)
plt.grid()
#plt.savefig('control_varying_delta_sum_Pmax_0.1.png', bbox_inches='tight', dpi = 80)
plt.savefig('control_varying_delta_sum_Pmax_01.png', bbox_inches='tight', dpi = 40)
plt.show()
```

```python
df3 = pd.read_csv('Data/control_delta_Pmax_both_varying', delim_whitespace = True)
df3.head()
```

```python
delta3 = df3.iloc[:,0]
Pmax3 = df3.iloc[:,1]
controls3 = df3.iloc[:, 2:9]
controls_abs3 = controls3.abs()
controls_sum3 = controls_abs3.sum(axis=1)
```

```python
delta3 = np.matrix(delta3)
Pmax3 = np.matrix(Pmax3)
control_sum3 = np.matrix(controls_sum3)
print(delta3.shape)
x = np.reshape(delta3, (492))
print(x.shape)
x = np.reshape(delta3, (123, 4))
print(x.shape)
```

```python
len(delta3)
x = np.reshape(delta3, (4, 123))
y = np.reshape(Pmax3, (4, 123))
z = np.reshape(control_sum3, (4, 123))
```

```python
fig = plt.figure(figsize=(14,12))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z)

ax.set_xlabel('$\delta$', fontsize=22, rotation=150, size = 26)
ax.set_ylabel('$Pmax$', fontsize=20, size = 26)
ax.set_zlabel('$\sum_i | P_{max} tanh( \delta \cdot \Theta_{i} )| $', fontsize=20, rotation=60, size = 26)
plt.savefig('surface.png', bbox_inches='tight', dpi = 50)
plt.show()
```
